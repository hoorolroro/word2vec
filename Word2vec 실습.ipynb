```
model = KeyedVectors.load_word2vec_format("./GoogleNews-vectors-negative300.bin", 
                                          binary=True,
                                          limit=60000)
```

```
print(model['apple'])
```

# Similarity

```
print("similarity between Benz and BMW: {}".format(model.similarity("Benz", "BMW")))
print("similarity between apple and banana: {}".format(model.similarity("apple", "banana")))
print("similarity between kid and adult: {}".format(model.similarity("kid", "adult")))
print("similarity between math and science: {}".format(model.similarity("math", "science")))
print("similarity between rice and chicken: {}".format(model.similarity("rice", "chicken")))
print("similarity between love and friendship: {}".format(model.similarity("love", "friendship")))
print("similarity between gun and terror: {}".format(model.similarity("gun", "terror")))
print("similarity between heart and body: {}".format(model.similarity("heart", "body")))
print("similarity between book and knowledge: {}".format(model.similarity("book", "knowledge")))
print("similarity between ocean and blue: {}".format(model.similarity("ocean", "blue")))
print(model.most_similar("ocean", topn=10))
```

# Word Anology

```
print(model.most_similar(positive=['king', 'women'], negative=['man'], topn=1))
print(model.most_similar(positive=['high', 'school'], negative=['high'], topn=1))
print(model.most_similar(positive=['congressman', 'woman'], negative=['man'], topn=1))
print(model.most_similar(positive=['beautiful', 'woman'], negative=['beautiful'], topn=1))
print(model.most_similar(positive=['guitar', 'player'], negative=['guitar'], topn=1))
print(model.most_similar(positive=['nice', 'guy'], negative=['bad'], topn=1))
print(model.most_similar(positive=['polite', 'man'], negative=['rude'], topn=1))
print(model.most_similar(positive=['nothing', 'better'], negative=['worse'], topn=1))
print(model.most_similar(positive=['god', 'woman'], negative=['man'], topn=1))
print(model.most_similar(negative=['king', 'women'], positive=['soldier'], topn=1))
```

```
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence

sentences = LineSentence("news1.txt")
model = Word2Vec(sentences, size=100, window=5, min_count=1, workers=4)
model.save("word2vec.model")
```

```
model = Word2Vec.load("word2vec.model")
print(model.wv.most_similar("car", topn=200))
print(len(model.wv.vocab))
```
